```python
from loki import ir
from loki import FindNodes
from loki import Scalar
from loki import Assignment
#remove JLON loops
loop_map= {}
routine_new=routine.clone()
for loop in FindNodes(ir.Loop).visit(routine_new.body):
    if loop.variable==horizontal.index:
        loop_map[loop]=loop.body
    
routine_new.body=Transformer(loop_map).visit(routine.body)

#change the name of the routine
#routine.spec.append(ir.Pragma(keyword='acc', content='routine seq'))
print("=================================")
print("=================================")
routine_new.name=routine_new.name+'_OPENACC'
print(routine_new.name)
print("=================================")
print("=================================")

#Add acc routine (routine) seq 
#routine.spec.insert(0,ir.Pragma(keyword='acc', content='routine seq'))
routine_new.spec.insert(0,ir.Pragma(keyword='acc', content='routine ('+routine_new.name+') seq'))
routine_new.spec.insert(1,ir.Comment(text=''))

#Add JLON=KIDIA
jlon=Scalar("JLON")
kidia=Scalar("KIDIA")
routine_new.spec.append(Assignment(jlon, kidia))


#from loki import fgen
#print(fgen(routine.spec))

#we want to add USE STACK_MOD and #include stack.h statement just above IMPLICIT NONE
idx=0
for spec in routine_new.spec.body:
    if type(spec)==ir.Intrinsic and spec.text=='IMPLICIT NONE': #find IMPLCICIT NONE location in the routine spec
        break
    idx=idx+1
routine_new.spec.insert(idx,ir.Import(module='STACK_MOD'))
routine_new.spec.insert(idx+1,ir.Import(module='stack.h',c_import=True))
routine_new.spec.insert(idx+2,ir.Comment(text=''))


from loki import fgen
print(fgen(routine.spec))
routine=routine_new.clone()
```

In this notebook, we will not go into detail on how to actually modify the control flow tree of this routine. But we will extract this routine from the module and put it into a standalone module.

Let's start by creating a clone of this routine with a new name:

```python
my_routine = lite_loop.clone(name='my_routine')
my_routine
```

Next, we create a new module and insert `my_routine` as a subroutine. To make sure the relevant declarations from the original module are available, we create a copy of the relevant spec:

```python
from loki import Module
my_module = Module(name='my_module', spec=phys_mod.spec.clone(), contains=(my_routine,))
my_module
```

Let's ensure the new module contains `my_routine`:

```python
assert len(my_module.subroutines) == 1
assert my_module.subroutines[0] is my_routine
```

We can also take a look at the Fortran code of this new module:

```python
print(my_module.to_fortran())
```

And, ultimately, we can write this to a separate source file:

```python
from pathlib import Path
Sourcefile.to_file(fgen(my_module), Path('my_module.F90'))
```
